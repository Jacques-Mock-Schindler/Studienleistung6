```python
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib.ticker import FuncFormatter
```

```python
x = [i for i in range(1, 10001)]
y_lin = [i for i in x]
```

```python
y_log = [math.log2(i) for i in x]
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

```python
plt.plot(x, y_lin, color="red", label="$f(x)=x$")
plt.plot(x, y_log, color="green", label="$g(x)=log_2 x$")
plt.suptitle("Lineares und logarithmisches Wachstum")
plt.title("(Gegenüberstellung)")
plt.xlabel("Anzahl Elemente in der Suchsequenz")
plt.ylabel("Maximal erforderliche Vergleiche")
plt.legend(frameon=False, loc="upper left")
plt.savefig("lin_log.svg")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter

# Beispieldaten erstellen
x = np.arange(0, 100)
y = x ** 2

# Plot erstellen
plt.figure(figsize=(10, 6))
plt.plot(x, y)

# Tausendertrennzeichen für die Y-Achse einstellen
formatter = FuncFormatter(lambda x, _: f'{x:,.0f}')
plt.gca().yaxis.set_major_formatter(formatter)

plt.title("Plot mit Tausendertrennzeichen")
plt.xlabel("X-Achse")
plt.ylabel("Y-Achse")

plt.show()
```
